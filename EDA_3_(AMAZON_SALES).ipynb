10 . ' Identify any potential areas for improvement or optimization based on the data analysi ?

1. Low Rating Categories
Categories like Car & Motorbike and Musical Instruments have relatively lower average ratings.

- Action: Improve product quality or customer support in these categories to enhance customer satisfaction.

2. High Price, Low Ratings
Some products with higher discounted prices are not rated better than cheaper products.

- Action: Brands should ensure that premium pricing matches customer expectations (better quality, packaging, durability).

3. Unoptimized Discounts
While some categories offer consistently high discounts, their rating impact is low.

- Action: Re-evaluate discount strategies  rather than big discounts, focus on value-added features.

4. Generic Product Titles
Many product names use repetitive keywords like "fast", "charging", "braided", which may cause confusion.

- Action: Use unique and detailed product titles for better product visibility and searchability.


5. Customer Review Keywords
Reviews indicate frequent mentions of quality concerns like "not durable", "low performance", especially in tech accessories.

- Action: Use this feedback to identify specific quality issues and optimize product design or manufacturing.

6. Top Categories  Further Optimization
Categories like Office Products and Toys & Games have high satisfaction and can be further promoted.

- Action: Focus marketing and cross-selling strategies on these high-performing categories.
